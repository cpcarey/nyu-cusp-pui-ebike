This notebook simply reads and displays the available shapefiles for visual validation.

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
zips_sfba = gpd.read_file('../data/shapefiles/zipcodes/san_francisco_bay_area.shp')
zips_sf = gpd.read_file('../data/shapefiles/zipcodes/san_francisco.shp')
zips_dc = gpd.read_file('../data/shapefiles/zipcodes/washington_dc.shp')
```

```python
def plot_zipcodes(gdf):
    fig, ax = plt.subplots(figsize=(8, 8))
    gdf.plot(ax=ax)
    plt.show()
```

```python
zipcode_gdfs = [zips_sfba, zips_sf, zips_dc]
for gdf in zipcode_gdfs:
    plot_zipcodes(gdf)
```
