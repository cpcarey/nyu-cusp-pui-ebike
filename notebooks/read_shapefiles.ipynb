This notebook simply reads and displays the available shapefiles for visual validation.

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
ZIPCODE_DIR = '../data/shapefiles/zipcodes'
zips_sfba = gpd.read_file('{}/san_francisco_bay_area.shp'.format(ZIPCODE_DIR))
zips_sf = gpd.read_file('{}/san_francisco.shp'.format(ZIPCODE_DIR))
zips_dc = gpd.read_file('{}/washington_dc.shp'.format(ZIPCODE_DIR))
```

```python
def plot_zipcodes(gdf):
    fig, ax = plt.subplots(figsize=(8, 8))
    gdf.plot(ax=ax)
    plt.show()
```

```python
zipcode_gdfs = [zips_sfba, zips_sf, zips_dc]
for gdf in zipcode_gdfs:
    plot_zipcodes(gdf)
```
