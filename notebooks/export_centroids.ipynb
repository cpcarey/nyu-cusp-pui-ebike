```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from functools import reduce
from geopy import Point
from geopy import distance

import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
EXPORTS_DIR = '../data/exports'
OUTPUT_DIR = '../data/exports'

K = 9 
```

```python
class AnalysisConfig:

    def __init__(self,
                 export_path='',
                 labels_path='',
                 stations_path=''):
        self.export_path = export_path
        self.labels_path = labels_path
        self.stations_path = stations_path
```

```python
config_sf = AnalysisConfig(
    export_path=f'{OUTPUT_DIR}/centroids_k{K}_sf.csv',
    labels_path=f'{EXPORTS_DIR}/cluster_labels_{K}_sf.csv',
    stations_path=f'{EXPORTS_DIR}/SF_ele_single station.csv',
)

config_dc = AnalysisConfig(
    export_path=f'{OUTPUT_DIR}/centroids_k{K}_dc.csv',
    labels_path=f'{EXPORTS_DIR}/cluster_labels_{K}_dc.csv',
    stations_path=f'{EXPORTS_DIR}/DC_ele_single station.csv',
)
```

```python
config = config_sf
```

```python
stations_df = pd.read_csv(config.stations_path, index_col=0)
labels_df = pd.read_csv(config.labels_path, index_col=0)
labels_df = labels_df.join(stations_df)
labels_df[['count']] = labels_df[['label']].copy() * 0
labels_df = labels_df.set_index('label')
display(labels_df)
```

```python
labels_agg_df = labels_df.groupby(['label']).agg({
    'lat': 'mean',
    'lng': 'mean',
    'elevation': 'mean',
    'count': 'count',
})
display(labels_agg_df.head())
```

```python
labels_agg_df.to_csv(config.export_path, index=False)
```
